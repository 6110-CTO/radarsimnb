# Receiver Operating Characteristic (ROC)

```python
from radarsimpy import roc_pd, roc_snr
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
sns.set_style('darkgrid')
```

```python
snr = np.arange(-10, 30, 0.1)
pfa = np.array([1e-4, 1e-6, 1e-8])
pd = roc_pd(pfa, snr, 1, 'Swerling 3')

it_pfa = np.nditer(pfa, flags=['f_index'])
while not it_pfa.finished:
    plt.plot(snr, pd[it_pfa.index, :], label='Pfa = '+str(it_pfa[0])+', N = 1')
    it_pfa.iternext()


pd = roc_pd(pfa, snr, 20, 'Swerling 3')

it_pfa = np.nditer(pfa, flags=['f_index'])
while not it_pfa.finished:
    plt.plot(snr, pd[it_pfa.index, :],
             label='Pfa = '+str(it_pfa[0])+', N = 20')
    it_pfa.iternext()

plt.legend()
plt.xlabel('SNR (dB)')
plt.ylabel('Probability of detection')
```

```python
snr = np.arange(-10, 30, 0.1)
pfa = 1e-6
N = 5
pd_sw1 = roc_pd(pfa, snr, N, 'Swerling 1')
pd_sw2 = roc_pd(pfa, snr, N, 'Swerling 2')
pd_sw3 = roc_pd(pfa, snr, N, 'Swerling 3')
pd_sw4 = roc_pd(pfa, snr, N, 'Swerling 4')
pd_sw5 = roc_pd(pfa, snr, N, 'Swerling 5')
pd_coherent = roc_pd(pfa, snr, N, 'Coherent')
pd_real = roc_pd(pfa, snr, N, 'Real')

plt.plot(snr, pd_sw1, label='Swerling 1'+', N = '+str(N))
plt.plot(snr, pd_sw2, label='Swerling 2'+', N = '+str(N))
plt.plot(snr, pd_sw3, label='Swerling 3'+', N = '+str(N))
plt.plot(snr, pd_sw4, label='Swerling 4'+', N = '+str(N))
plt.plot(snr, pd_sw5, label='Swerling 5'+', N = '+str(N))
plt.plot(snr, pd_coherent, label='Coherent'+', N = '+str(N))
plt.plot(snr, pd_real, label='Real'+', N = '+str(N))

plt.legend()
plt.xlabel('SNR (dB)')
plt.ylabel('Probability of detection')
```

```python
pfa = 1e-5
pd = np.arange(0.1, 0.9, 0.01)
N = 10
snr_real = roc_snr(1e-4, pd, N, 'Real')

snr_sw1 = roc_snr(pfa, pd, N, 'Swerling 1')
snr_sw2 = roc_snr(pfa, pd, N, 'Swerling 2')
snr_sw3 = roc_snr(pfa, pd, N, 'Swerling 3')
snr_sw4 = roc_snr(pfa, pd, N, 'Swerling 4')
snr_sw5 = roc_snr(pfa, pd, N, 'Swerling 5')
snr_coherent = roc_snr(pfa, pd, N, 'Coherent')
snr_real = roc_snr(pfa, pd, N, 'Real')


plt.plot(pd, snr_sw1, label='Swerling 1'+', N = '+str(N))
plt.plot(pd, snr_sw2, label='Swerling 2'+', N = '+str(N))
plt.plot(pd, snr_sw3, label='Swerling 3'+', N = '+str(N))
plt.plot(pd, snr_sw4, label='Swerling 4'+', N = '+str(N))
plt.plot(pd, snr_sw5, label='Swerling 5'+', N = '+str(N))
plt.plot(pd, snr_coherent, label='Coherent'+', N = '+str(N))
plt.plot(pd, snr_real, label='Real'+', N = '+str(N))

plt.legend()
plt.xlabel('Probability of detection')
plt.ylabel('SNR (dB)')
```
